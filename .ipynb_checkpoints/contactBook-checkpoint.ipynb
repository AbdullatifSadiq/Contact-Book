```python
# Steps
# create contact list
# create contact
# Add contact to list
# Update contact list
# Delete contact

import json

# Saving contacts in a Json file.
def save_contact(contacts):
    with open('contact.json', 'w') as file:
        json.dump(contacts,file)

# Loading the saved contacts from the json file
def load_contact():
   try: 
       with open('contact.json', 'r') as file:
        return json.load(file)
       
       #handling errors for a missing file or an empty file
   except (FileNotFoundError,json.JSONDecodeError):
       return []
   
class Contact:

    #creating a contact class that would be called to create new contact instances. 
    def __init__(self,name,number,email):
        self.name = name
        self.number = number
        self.email = email
    
    #creating a dictionary that would contain contact details of a particular contact
    def contact_details(self):
        con_dict = {'Name': self.name, 'Number': self.number, 'Email': self.email}
        return con_dict
    
    #Adding a conatct to the contact list
    @staticmethod
    def add_contact(con,con_list):
        con_details = con.contact_details()
        con_list.append(con_details)
        print("Contact added successfully")
        save_contact(con_list)

        # Displaying the Contact Names with index
        print('\nContact List')
        for i, contact in enumerate(con_list):
          print(f"{i + 1}. {contact['Name']}")

    def find_contact(name_to_find,contact_list):
        found_contact = None
        for contact in contact_list:
            if contact['Name'].lower() == name_to_find.lower():
                found_contact = contact
                break
        
        if found_contact:
            print('Contact found') 
            for key, value in found_contact.items():
             print(f"{key}: {value}")
        else:
            print(f"No contact found with the name '{name_to_find}")

    def update_contact(name_to_update,contact_list):
        contact_to_update = None
        for contact in contact_list:
          if contact['Name'].lower() == name_to_update.lower():
            contact_to_update = contact
            break
        if contact_to_update:
            print('\nchoose an option')
            print('1. Phone number')
            print('2. Email Address')

            phone_or_email = int(input('What do you wish to update, Phone Number of Email'))
            if phone_or_email == 1:
                new_phone = input('Enter new phone number')
                contact_to_update['Number'] = new_phone
                print('Phone number updated successfully')

            elif phone_or_email == 2:
                new_email = input('Enter the new Email')
                contact_to_update['Email'] = new_email
                print('Email Address updated successfully')
        save_contact(contact_list)

            
    def delete_contact(index,con_list):
        if 0<= index < len(con_list):
            del con_list[index]
            print('Contact deleted succesfully')
        else:
            print('Invalid index. No Contact Deleted')

         # Displaying the updated Contact Names with index
        print('\nContact List')
        for i, contact in enumerate(con_list):
            print(f"{i + 1}. {contact['Name']}")

        save_contact(con_list)
    
contact_list = load_contact()

#Giving the user multiple options to choose from
print('\nchoose an option')
print('1. Create a new contact')
print('2. Find a Contact')
print('3. Update a Contact')
print('4. Delete a Contact')
print('5. Show Contact list')
print('6. Exit')

choice = int(input('Enter option 1,2,3,4, or 6'))

if choice == 1:
        #Taking contact details from user
        name = str(input('Enter contact name'))
        number = int(input('Enter phone number'))
        email = str(input('Enter contact email'))
        
        # Creating an instance of the 'Contact' class using the details from the user
        con = Contact(name,number,email)
        
        # Displaying the contact list containing the contact details
        Contact.add_contact(con,contact_list)

elif choice == 2:
    name_to_find = input('Enter the name of the contact you want to find')
    Contact.find_contact(name_to_find,contact_list)

elif choice == 3:
     name_to_update = input("Enter the name of the contact you want to update")
     Contact.update_contact(name_to_update,contact_list)


elif choice == 4:
     # Displaying the Contact Names with index
    print('\nContact List')
    for i, contact in enumerate(contact_list):
      print(f"{i + 1}. {contact['Name']}")

    # Asking the user for the index of the conatct to delete
    index_to_delete = input('Enter the index of the contact you want to delete')

    if index_to_delete.isdigit():
             index_to_delete = int(index_to_delete) - 1
             Contact.delete_contact(index_to_delete,contact_list)

elif choice == 5:
     # Displaying the Contact Names with index
        print('\nContact List')
        for i, contact in enumerate(contact_list):
          print(f"{i + 1}. {contact['Name']}")

elif choice == 6:
     print('Program Exited.')
         
else :
     print('Invalid choice. Enter 1,2,3,4 or 5')









```
